```python
## non linear  graph
#v=a*t
t=list(range(1,10))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(v,t)
plt.show()
```

```python
import numpy as np
t=np.arange(0.1,2.0,0.1)
v=[np.sin(i)for i in t]
plt.plot(v,t)
plt.show()
```
